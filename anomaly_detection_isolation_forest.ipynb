```python
import os
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, accuracy_score
```

```python
X_train = combined_df_train[['x', 'y', 'z', '010-000-024-033', '010-000-030-096', '020-000-032-221', '020-000-033-111']]
y_train = combined_df_train['anomaly']

X_test = combined_df_test[['x', 'y', 'z', '010-000-024-033', '010-000-030-096', '020-000-032-221', '020-000-033-111']]
y_test = combined_df_test['anomaly']


# Isolation Forest  
model = IsolationForest(n_estimators=100, contamination=0.04, random_state=11)
model.fit(X_train)
```

# contamination kritik      


```python

y_train_pred = model.predict(X_train)
y_train_pred = [1 if pred == -1 else 0 for pred in y_train_pred]
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")


y_test_pred = model.predict(X_test)
y_test_pred = [1 if pred == -1 else 0 for pred in y_test_pred]
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {test_accuracy:.4f}")

print(classification_report(y_test, y_test_pred))
```
