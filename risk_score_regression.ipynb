```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
 
data=pd.read_csv("Loan.csv")
data.head()
```

```python
data.columns
```

```python
data=data.drop(["ApplicationDate"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data["LoanPurpose"]
```

```python
categorical_columns = ['EmploymentStatus', 'EducationLevel', 'MaritalStatus', 
                       'HomeOwnershipStatus', 'LoanPurpose']

data_encoded = pd.get_dummies(data, columns=categorical_columns)
print(data_encoded)
```

```python
data_encoded
```

```python
data_encoded.columns
```

```python
data_encoded 
```

```python
data_encoded
```

```python
X = data_encoded.drop(['RiskScore'], axis=1)   
y = data_encoded['RiskScore']   
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
(X_train.shape[1],)
```

```python
## Set up the Tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## Set up Early Stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),  
    tf.keras.layers.Dense(32, activation='relu'),  
    tf.keras.layers.Dense(1)  
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error', 'mean_squared_error']) # Regresyon için MSE kullanılır
history = model.fit(X_train_scaled, y_train, epochs=40, batch_size=32, validation_data=(X_test_scaled, y_test))


test_loss = model.evaluate(X_test_scaled, y_test)
print(f"Test Loss: {test_loss}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

# Tahminlerinizi ve gerçek veriyi alarak performansı ölçme
y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
# Yeni verilerle tahmin yapma
y_pred = model.predict(X_test_scaled)

# Sonuçları karşılaştırmak için gerçek ve tahmin edilen değerler
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred.flatten()})
print(comparison.head())
```
