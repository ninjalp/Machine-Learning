```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples=np.load("samples.npy")

X=samples[:,0]
r=samples[:,1]
plt.figure()
plt.plot(X,"green")
plt.plot(r,"black")
plt.grid
```

```python
X_mean=X.mean()
r_mean=r.mean()
```

```python
print("x'in ortalama değeri:",X_mean)
print("r nin ortalama değeri:",r_mean)
```

```python
g=100
üst_taraf=[]
alt_taraf=[]
for i in range(g): 
    üst_taraf.append((X[i]-X_mean) * (r[i]-r_mean))
for i in range(g):
    alt_taraf.append((X[i]-X_mean)**2)
üst_taraf_numpy=np.array(üst_taraf)
alt_taraf_numpy=np.array(alt_taraf)
w1=üst_taraf_numpy.sum()/alt_taraf_numpy.sum()
w0=r_mean-(w1*X_mean)
print("w1 in değeri:",w1)
print("w0 ın değeri:",w0)
```

```python
rnew=w0+w1*X
print("regresyon doğrusu değerleri:",rnew)
```

```python
plt.figure(figsize=(12,12))
figure1=plt.plot(X,r,"*")
figure2=plt.plot(X,rnew)
plt.grid()
```
