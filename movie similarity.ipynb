```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
for index, row in movieRatings.iterrows():
    count = row.count()
    if count > 1:
        print(f"User {index} has rated {count} movies.")
```

```python
counts = movieRatings.count()

# Find the column with the highest count
most_rated_movie = counts.idxmax()

print(f"The most rated movie is {most_rated_movie} with {counts[most_rated_movie]} ratings.")
```

```python


# Calculate the mean rating for each movie
mean_ratings =movieRatings.mean()

# Print the top 10 movies with the highest mean rating
top_10_movies = mean_ratings.nlargest(10)
print("Top 10 movies by mean rating:")
for movie, rating in top_10_movies.items():
    print(f"{movie}: {rating:.2f}")
```

```python
import pandas as pd



# Count the number of non-NaN values for each column (i.e. number of ratings for each movie)
movie_counts = movieRatings.count()

# Sort the movie counts in descending order and select the top 10
top_10_movies = movie_counts.sort_values(ascending=False).head(10)

# Print the top 10 most rated movies
print("Top 10 most rated movies:")
for movie, count in top_10_movies.items():
    print(f"{movie}: {count} ratings")
```
