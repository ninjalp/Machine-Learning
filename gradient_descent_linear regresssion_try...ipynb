```python
X_mean=x.mean()
y_mean=y.mean()
print(X_mean)
print(y_mean)
```

```python
X=x

g=100

st_taraf=[]
alt_taraf=[]
for i in range(g): 
    st_taraf.append((X[i]-X_mean) * (y[i]-y_mean))
for i in range(g):
    alt_taraf.append((X[i]-X_mean)**2)
st_taraf_numpy=np.array(st_taraf)
alt_taraf_numpy=np.array(alt_taraf)
w1=st_taraf_numpy.sum()/alt_taraf_numpy.sum()
```

```python
w0=w0=y_mean-(w1*X_mean)
ynew=w0+w1*x
plt.figure(figsize=(15,15))
figure3=plt.scatter(x,y,color="blue")
figure4=plt.plot(X,ynew,"red")
plt.grid()
```

# Gradient descent

```python

m=0
c=0
activation_function= 1/(1 + np.exp(-x)) #aktivasyon fonksiyonu,lerning rate.
iteration=1000000
n=len(x)

#mse = ((y-(m*x+c))**2).sum()//len(x) bunun trevini alcaz bir elimizde  ; parametreler (m,c) m iin trev aldmzda trev:
#(-2/n) * ((x *(y-y_pred)).sum())
 
    
    #c iin trev aldmzda :
#(-2/n) * ((y-y_pred).sum())

for i in range(iteration):
    y_pred=m*x+c
    m_trevi=(-2/n) * ((x *(y-y_pred)).sum())
    c_trevi=(-2/n) * ((y-y_pred).sum())
    m=m-activation_function*m_trevi#yeni m
    c=c-activation_function*c_trevi#yeni c
m
```

```python
c
```

```python
Y_pred = m*x + c
plt.figure(figsize=(15,15))
plt.scatter(X, y) 
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)], color='red')  
plt.grid()
```
